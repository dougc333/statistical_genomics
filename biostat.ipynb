<p>Statistical power is the probability of the hypothesis test detecting an effect; if there is one. This 
analysis can be used to find the minimum sample size N. The null hypothesis test makes an assumption
about the outcome. The output of the test is a p-valu which is the probability of observing
the result if the null hypothesis is true. You pick a significance level to determing statistical significane.
If the result p-value is less than the statistical significane p-value then teh null hyphthesis is accepted. if the 
calculated p-value is greather than the statistical significance p-value then teh nuyll hypothesis is rejected. </p>
<p>The 4 related parameters are 1) Effect Size,the result, ie pearsons correlation coefficient for the relationship
between 2 variables. 2) Sample Size,N 3) Significance, p value to define statistical significance
usually set to .05 or 5%. 4) Statistical power, usually set
at 80%. These 4 variables are all related. A larger N means the effect can be easier to detect(this is true? how about
bonferroni adjusting the p by p/N? statistical power can be increased by decreasing the signifcance level. </p>
<p>Example:Student t-test for comparing if the means of 2 samples are identical even though the mean is not 
numerically identical. ie. is a mean of 3.4 the same as mean of 4? p_alpha=0.05, effect_size=.80, power = 0.8.
With these defaults what the number of observations from each sample/batch to detect an effect of 0.8 with 
an 80% chance of detecting the effect if true? 20% for type2 error, 5% if there is no such effect or type 1 error. 
False negative rate is type 2 error? Yes, type 1 error is false positive. </p>

```python
<h6>week1 confounding and randomization</h6>
<p>The solution to the confounding problem is randomization</p>
<p>Confounding is when you are looking at the relationship between 2 variables and there is another 
hidden variable affecting the output. This hidden variable can affect results. 
If we are looking for an association between gene expression levels and
control/disease group or phenotype there may be other variables affecting this relationship. Another var
is a confounding variable. Example kid and adult there is an output, literate, adult is literate kid is not. And
you can correlate show size, kid small, adult large so does large shoe size mean literate? </p>
<img src="week1/week1_2a.png">
<p>Confounding example: there is an intermediate variable age which would model the relationship of literacy. The confounding variable is related to 
both age and literacy. JL compares a literate adult to an illiterate child. The input can be shoe size and that
is correlated to litteracy levels. But the hidden variable is really number of years of education or age assuming age
correlates with years of education</p>
<img src="week1/week1_2b.png">
<p>GSEA example on confounding variable problem: 78% differentially expressed when you look at asian vs eur but 
    96% diff. expresseed when you look at year. So year was a more important predictor. And this study was fishy
    because past studies showed no correlations between ethnicity and expression. Turns out the data was collected
    using different methods, ie microarray vs. sequencing and so variations were due to technology variation
    not from the ethnicity or time variable! </p>
<img src="week1/week1_2c.png">
<p>note the clustering on years, time differences from technology variation not genetic variation! </p>
<img src="week1/week1_2d.png">
<p>When you adjust for years you see no statistically significant effect! How do you do the correction?</p>
<img src="week1/week1_2e.png">
<p>study showed people >100 had genetic variants that caused longevity. turns out this was an error 
in technology variation bc samples on young people were collected on 1 technology and the samples on 
older were collected on another tech. </p>
<img src="week1/week1_2f.png">
<p>occurs in proteomics; correlated a protemic variation to cancer but turns out there was a batch effect
related to time so you were measuring time variation vs. biology variation</p>
<img src="week1/week1_2g.png">
<p>paper worth reading talks about batch effects. Batch effects include operator, reagent biases, high throughput
systems can remove these, low throughput cant. Normaalization is common for GSEA,and is only 1 technique. 
ie normalization across microarrray and 
sequender output. Normalization reduces variance w/o changing bias if done correctly. Normalization has 2 components
1) scale intensities so each method/batch has same value. 2) transorm data using (x-mu)/sigma so distribution of probe intensities
is same on all samples is the simplistic view. Not what is done in practice. 
</p>
Batch effects is complicated field, open source sw ranging from PCA to canonical correlation analysis, etc.. many techniques
one simplistic view is a tsne plot of batch corrected data. Looks better in that visually batch effect isnt adding 
confounding variables to the data. But still needs some work to make into system heuristics for quality. 
<img src="week1/week1_2h.png">
<img src="week1/week1_2i.png">
<img src="week1/week1_2j.png">
<h6>Adding randomization to confounding variable</h6>
<p>Things you can do at the experiment design stage for batch effects. 
dark means confounding variable is high, light means low. When you do the experiment, the red circles are 
the samples in 1 treatment batch and the green another treatment batch so you have measured the confounding variable and not the 
real gene expression. </p>
<img src="week1/week1_2k.png">
<p>another way to design the experiment for each patient you randomly assign a green or red treatment.</p>
<img src="week1/week1_2l.png">
<p>Another design method: blocking</p>
<img src="week1/week1_2m.png">
<p>If run all female Control in week 1 and all male Treated in week2 hard to distinguish between 
male/femake effect, week1, week2 effect, control/disease effect. </p>
<img src="week1/week1_2n.png">
<p>block the subjects to get equal balance</p>
<img src="week1/week1_2o.png">
<p>control, balanced means same yellow/green in each slot. Lot more info on experiment design, balanced replicatin, controls.
Q: how do you know from data if somethign is balanced? Uneven variances, uneven N. </p>
<img src="week1/week1_2p.png">
```
